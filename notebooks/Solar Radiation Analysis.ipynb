```python
# Removing (dropping) the comment columns because it is empty in all datasets
Benin_data = Benin_data.drop(columns=['Comments'])
Sierraleone_data = Sierraleone_data.drop(columns=['Comments'])
Togo_data = Togo_data.drop(columns=['Comments'])
```

```python
## Adding country column to each dataset for later identification
Benin_data['Country']="Benin"
Sierraleone_data['Country']="Sierraleone"
Togo_data['Country']="Togo"
```

```python
# Checking if the Country Column is added
print(Benin_data.columns)
print(Sierraleone_data.columns)
print(Togo_data.columns)
```

# Combining the dataset after adding the country column to each data

```python
# Combining the datasets of the three countries
data = pd.concat([Benin_data, Sierraleone_data, Togo_data])
```

## 2. Data Quality Checking

```python
# Chekin for Missing Values
print(data.isnull().sum())
```

```python
data.duplicated()
```

```python
# Chekin for invalid input e.g, negative values where only positive should exist for GHI, DNI, DHI
print(data[(data['GHI'] < 0) | (data['DNI'] < 0) | (data['DHI'] < 0)])
```

```python
# Outlier Detection using IQR for ModA
Q1 = data['ModA'].quantile(0.25)
Q3 = data['ModA'].quantile(0.75)
IQR = Q3 - Q1
outlier_threshold = 1.5 * IQR
outliers = data[(data['ModA'] < Q1 - outlier_threshold) | (data['ModA'] > Q3 + outlier_threshold)]
print(outliers) 
```

```python
 # Create a box plot for the "Price" attribute
 plt.figure(figsize=(10, 6))
 plt.boxplot(data['ModA'], vert=True, 
patch_artist=True)
 plt.title('Box Plot of ModA')
 plt.xlabel('ModA')
 plt.show()
```

```python
# List of numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Loop through numeric columns to plot boxplots grouped by 'Country'
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    data.boxplot(column=col, by='Country', grid=False, patch_artist=True)
    plt.title(f'Box Plot of {col} by Country')
    plt.suptitle('')  # Suppress the automatic suptitle from pandas boxplot
    plt.xlabel('Country')
    plt.ylabel(col)
    plt.show()
```
